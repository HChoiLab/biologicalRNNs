#### Generate sampled datasets and save

```python
seq_len = dff_pyr_visl_L23.shape[-1] -1
```

```python
def generate_random_averages(arr, num_samples=2000, num_neurons=100, filtering = True, sigma = 1):
    n, seqLen_full = arr.shape
    seqLen = seqLen_full-1
    
    arr_train = np.zeros((num_samples,seqLen))
    arr_test = np.zeros((num_samples,seqLen))
    
    idxs = np.random.choice(n, (num_samples,num_neurons), replace=True)
    
    for ii in range(num_samples):
        ## Predictive fitting
        arr_train[ii] = np.mean(arr[idxs[ii]],0)[:seqLen]
        arr_test[ii] = np.mean(arr[idxs[ii]],0)[1:]
        ## copy recon
#         arr_train[ii] = np.mean(arr[idxs[ii]],0)[:seqLen]
#         arr_test[ii] = np.mean(arr[idxs[ii]],0)[0:]

    if filtering:
        arr_train = gaussian_filter1d(arr_train, sigma)
        arr_test = gaussian_filter1d(arr_test, sigma)

    return arr_train, arr_test
```

```python
dff_V1_L4_pyr_train, dff_V1_L4_pyr_test = generate_random_averages(dff_pyr_visp_L4)
dff_V1_L4_sst_train, dff_V1_L4_sst_test = generate_random_averages(dff_sst_visp_L4)
dff_V1_L4_vip_train, dff_V1_L4_vip_test = generate_random_averages(dff_vip_visp_L4)

dff_V1_L23_pyr_train, dff_V1_L23_pyr_test = generate_random_averages(dff_pyr_visp_L23)
dff_V1_L23_sst_train, dff_V1_L23_sst_test = generate_random_averages(dff_sst_visp_L23)
dff_V1_L23_vip_train, dff_V1_L23_vip_test = generate_random_averages(dff_vip_visp_L23)

dff_V1_L5_pyr_train, dff_V1_L5_pyr_test = generate_random_averages(dff_pyr_visp_L5)
dff_V1_L5_sst_train, dff_V1_L5_sst_test = generate_random_averages(dff_sst_visp_L5)
dff_V1_L5_vip_train, dff_V1_L5_vip_test = generate_random_averages(dff_vip_visp_L5)
```
